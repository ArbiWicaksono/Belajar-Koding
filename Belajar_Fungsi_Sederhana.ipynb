Coba
f(x) = 2x jika 0<x<=10 dan
f(x) = sin(x) jika 10<x<=20 dengan incremenet 0.1


```python
x = 0.0
x_max = 20.0


print("x       y1 (2x)        y2 (sin(x))")
print("-----------------------------------")

while x <= x_max:
    if x <= 10:
        y1 = 2 * x
        print("{0:.1f}     {1:.3f}    -".format(x, y1))
    else:
        y2 = np.sin(x)
        print("{0:.1f}     -         {1:.3f}".format(x, y2))
    x += 0.1
```

Membuat Grafiknya dengan menggunakan sistem List

```python
#Inisialisasi List Kosong
x_values_y1 = []
y1_values = []
x_values_y2 = []
y2_values = []

#Loop While
x = 0.0
x_max= 20.0

while x <= x_max:
  if x <= 10:
    x_values_y1.append(x)
    y1_values.append(2*x)
  else:
    x_values_y2.append(x)
    y2_values.append(np.sin(x))

  x += 0.1

#Plotting
plt.figure(figsize=(10, 5))
plt.plot(x_values_y1, y1_values, label="y1 = 2x (x  10)", color='blue')
plt.plot(x_values_y2, y2_values, label="y2 = sin(x) (x > 10)", color='green')

plt.title("Grafik y1 = 2x dan y2 = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
